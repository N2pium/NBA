```python
import pickle
import pandas as pd

# Load model and predictors
with open("nba_model.pkl", "rb") as f:
    rr, sfs, predictors = pickle.load(f)

# Load the dataset to use for predictions
df = pd.read_csv("NBA_games.csv", index_col=0)
df = df.sort_values("date")
```

```python
full = pd.read_csv("NBA_games_merged.csv", index_col=0)
```

```python
def predict_matchup(team_a, team_b, full, model, predictors):
    # Get rows for each team
    t1 = full[full["team_x"] == team_a].iloc[-1]
    t2 = full[full["team_y"] == team_b].iloc[-1]

    # Make sure both exist
    if t1.empty or t2.empty:
        raise ValueError("One or both teams not found in merged dataset.")

    # Compute difference
    diff = (t1[predictors] - t2[predictors]).to_frame().T

    result = model.predict(diff)[0]
    winner = team_a if result == 1 else team_b
    print(f"Predicted winner: {winner}")
    return winner
```

```python
predict_matchup("IND", "HOU", full, rr, predictors) #boom shakalaka 
```
